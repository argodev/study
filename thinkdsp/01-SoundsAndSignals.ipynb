- "A signal is a quantity that varies in time" - in this model, there are tons of *signals* in the world around us...
- a *transducer* is a device that converts a signal from one form to another
- A signal can also be a value that varies in *space* (not just time). Think here of the elevation of a mountain trail
- Another thought might be a multi-dimensional signal
    - image varies in two-dimensional space
    - movie varies in three dimensions - 2d (image) and time

```python
from thinkdsp import CosSignal, SinSignal, decorate

# instantiate cosine and sine signals
cos_sig = CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = SinSignal(freq=880, amp=0.5, offset=0)

# plot the cosine signals. By default, `plot` plots three periods
cos_sig.plot()
decorate(xlabel='Time (s)')
```

```python
# and now the sine signal
sin_sig.plot()
decorate(xlabel='Time (s)')
```
